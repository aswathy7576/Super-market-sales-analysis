```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_excel('superstore_sales.xlsx')
df.head() 
```

#### Satır ve Sütun Sayısını Ögrenme - Get the number of rows and columns

```python
df.shape
```

#### Eksik veri olup olmadığını kontrol etme -Count the number of missing values


```python
df.isnull().sum()
```

<H4> En çok satan ürünler hangileri?  -  Which are the most selling products?


```python
boughts = df.groupby('product_name').agg({'quantity': 'sum'}).sort_values('quantity',ascending=False)[:10]
boughts
```

```python
sns.barplot(x='product_name',y='quantity',data=boughts.reset_index())
plt.xticks(rotation = 'vertical')
```

 <h4> Satışı en iyi 10 ürün hangileridir? - Which are the Top 10 products by sales?

```python
productSales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
productSales.sort_values(by=['sales'], inplace=True, ascending=False)
productSales.head(10)
```

<h4> Hangi ürün daha çok kar getirmiş ?  Which are the most profitable products?


```python
profitable_products = pd.DataFrame(df.groupby(['product_name']).sum()['profit'])
profitable_products.sort_values(by=['profit'], inplace = True, ascending = False)
profitable_products.head(10)
```

```python
sns.barplot(x='product_name',y='profit',data=profitable_products.head(10).reset_index())
plt.xticks(rotation = 'vertical')
```

<h4>Hangi kategori daha çok satış yapmış? -  What category sold the most?


```python
df.groupby(['category','sales']).sum()
```

<h4> Hangi kategori daha çok karlı? - Which are the most profitable category?

```python
profitable_category = pd.DataFrame(df.groupby(['category']).sum()['profit'])
profitable_category.sort_values(by=['profit'], inplace = True, ascending = False)
profitable_category
```

```python
plt.figure(figsize=(17, 5))
plt.bar(df['category'],df['sales'],color = 'b')
plt.xticks(size=8)
plt.show()
```

<h4> Kategori ve alt kategorideki toplam satış değerleri - Total sales values by category and subcategory


```python
df.groupby(['category','sub_category']).sum()
```

<h4> Alt kategorilerde en çok satan ürünler hangileri? - Which are the most selling products in subcategory? 

```python
df.groupby(["category", "sub_category"], as_index=False)["quantity"].count()
```

<h4> Hangi müşteri segmenti daha çok karlı ? - Which customer segments are the most profitable ?

```python
profitable_segment = pd.DataFrame(df.groupby(['segment']).sum()['profit'])
profitable_segment.sort_values(by=['profit'], inplace = True, ascending = False)
profitable_segment
```

```python
sns.barplot(x='segment',y='profit',data=profitable_segment.reset_index())
plt.xticks(rotation = 'vertical')
```

<h4> Kargo süreçlerinin satışa etkisi - What shipping modes sold the most products?

```python
plt.figure(figsize=(14, 6))
plt.bar(df['ship_mode'],df['sales'],color = 'b')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

#### Kargo süreçlerinin kategori bazlı satışa etkisi - Visualize the 'Category' column from the Shipmode column dataset standpoints.

```python
category_hist = sns.FacetGrid(df, col='ship_mode', palette='rainbow')
category_hist.map(plt.hist, 'category')
```

<h4> Hangi ürün pazarı daha çok satış yapıyor? What market sold the most products?

```python
sns.set_style('whitegrid')
sns.countplot(x='market',data=df, palette='rainbow')
```

<h4> En çok satış yapan ilk 10 ülke hangileridir ? - Which are the Top 10 country by sales? 

```python
countries = pd.DataFrame(df.groupby('country').sum()['sales'])
countries.sort_values('sales',inplace=True, ascending=False)
countries.head(10)
```

<h4> En çok satış yapan 10 ülkeyi içeren bir pasta grafik oluşturmak - Create a pie chart with the 10 countries that have the most sales

```python
countries = countries.sort_values(by = 'sales',ascending = False)[1:11]
countries['sales'].plot(kind='pie',autopct='%1.1f%%',figsize=(14,7))
plt.title('10 Countries with most sales')
plt.show()
```

<h4> Ortalama olarak en fazla kargo ücreti ödeyen ilk 10 ülkeler - Which are the average shipping cost for top 10 different countries?


```python
df.groupby('country').agg({'shipping_cost':'mean'}).sort_values('shipping_cost', ascending=False).head(10)
```

#### En çok kar getiren ilk 10 müşteri kimlerdir ? Who are the top-10 most profitable customers?


```python
data10 = df.sort_values('profit',ascending=False).head(10)[['order_id','customer_name','profit']]
data10
```

<h4> Sipariş tarihininden ay sutunu oluşturma - Add a ne column (month) 

```python
month = df['order_date'].astype(str) 
month = month.apply(lambda x:datetime.strptime(x,'%Y-%m-%d'))
df['month'] = month.map(lambda x: x.month) 
df
```

#### Yıl ve aylara göre toplam satış değerleri - Total sales values by year and month 

```python
date = df.groupby(['year','month']).sum()
date
```

#### Yıllara göre toplam satış grafiği - Total sales chart by the year 

```python
sns.barplot(x='year',y='sales',data=date.reset_index())
plt.xticks(rotation = 'vertical')
```
